```python
import geopandas as gpd

geology_map = gpd.read_file('/Users/simon/Work/Mars/SIM3292_MarsGlobalGeologicGIS_20M/SIM3292_Shapefiles/SIM3292_Global_Geology.shp')

geology_map
```

```python
geology_map.plot(column='Unit')
```

```python
geology_map.UnitDesc.value_counts()
```

```python
geology_map['Age'] = ""
geology_map['Age'].loc[geology_map['UnitDesc'].str.contains('Hesp')] = 'Hesperian'
geology_map['Age'].loc[geology_map['UnitDesc'].str.contains('Amaz')] = 'Amazonian'
geology_map['Age'].loc[geology_map['UnitDesc'].str.contains('Noach')] = 'Noachian'

geology_map
```

```python
simplified_map = geology_map.dissolve(by='Age')

simplified_map
```

```python
simplified_map.plot(column='Unit', legend=True)
```

```python
simplified_map.to_file('geology/simplified_geology.shp')
```
